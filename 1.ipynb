## Methods

```python
def operate(string: str, baseValue=0) -> int:
    if string[0] == '+':
        return baseValue + int(string[1:])
    else:
        return baseValue - int(string[1:])
```

```python
def procNumbers(string: str) -> int:
    total = 0
    listOfNumbers = string.split()
    for number in listOfNumbers:
        total = operate(number, total)
    return total
```

```python
def procNumbers2(string: str) -> int:
    total = 0
    previousResults = set({0})
    listOfNumbers = string.split()
    while True:
        for number in listOfNumbers:
            total = operate(number, total)
            if total not in previousResults:
                previousResults.add(total)
            else:
                return total
```
