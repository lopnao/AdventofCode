## Methods

```python
def countingRep(string:str)->int:
    boxIDs = string.split()
    countingDictTwo = set()
    countingDictThree = set()
    for boxID in boxIDs:
        for letter in set(boxID):
            if boxID.count(letter) == 2:
                countingDictTwo.add(boxID)
            elif boxID.count(letter) == 3:
                countingDictThree.add(boxID)
    return (len(countingDictTwo) * len(countingDictThree))
```

```python
def countDiff(string1:str,string2:str)->int:
    count = 0
    for index, letter in enumerate(string1):
        if letter != string2[index]:
            count += 1
    return count
```

```python
def testingIDs(string:str)->str:
    boxIDs = string.split()
    res = ''
    for boxID in boxIDs:
        for boxID2 in boxIDs:
            if countDiff(boxID, boxID2) == 1:
                for index, letter in enumerate(boxID):
                    if letter == boxID2[index]:
                        res += letter
                return res
    return None
```
