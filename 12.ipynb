```python
import getres
import collections
```

```python
data = getres.getInput(12).splitlines()
```

## Methods

```python
def getPots(data):
    pots = collections.deque([(p, v) for v, p in enumerate(data[0][15:])])
    if False in [pots[i][0] == '.' for i in range(5)]:
        [pots.appendleft(('.', -i)) for i in range(1, 6)]
    if False in [pots[len(pots) - 1 - i] == '.' for i in range(1, 6)]:
        [pots.append(('.', len(data[0][15:])+i)) for i in range(5)]
    return pots
```

```python
def getPatterns(data):
    return [d[:5] for d in data[2:] if d[-1] == '#'], [d[:5] for d in data[2:] if d[-1] == '.']
```

```python
def procGen(pots, patternTrue, patternFalse):
    newPosts = collections.deque(list(pots)[:])
    for v, pot in enumerate(list(pots)[2:len(pots)-3]):
        toTest = pots[0 + v][0] + pots[1 + v][0] + pots[2 + v][0] + pots[3 + v][0] + pots[4 + v][0]
        if toTest in patternTrue:newPosts[2 + v] = ('#', newPosts[2 + v][1])
        elif toTest in patternFalse:newPosts[2 + v] = ('.', newPosts[2 + v][1])
        else:newPosts[2 + v] = ('.', newPosts[2 + v][1])
    return newPosts
```

```python
def addIfNeed(pots):
    if False in [pots[i][0] == '.' for i in range(5)]:
        [pots.appendleft(('.', i)) for i in range(pots[0][1]-1, pots[0][1]-6, -1)]
    if False in [pots[len(pots) - 1 - i][0] == '.' for i in range(5)]:
        [pots.append(('.', i)) for i in range(pots[-1][1]+1, pots[-1][1]+6)]
    return pots
```

```python
def procData(data, nbOfGen=0):
    if nbOfGen > 200:
        return procData(data, 200) + (34 * (nbOfGen-200))
    pots = getPots(data)
    patternTrue, patternFalse = getPatterns(data)
    for i in range(nbOfGen):
        pots = procGen(pots, patternTrue, patternFalse)
        pots = addIfNeed(pots)
    return sum((elem[1] for elem in pots if elem[0] == '#'))
```

## Results

```python
for i in range(1, 300):
    a = procData(data, i)
    b = procData(data, i-1)
    print(f'Result : {a} and diff with previous is : {a - b} for i = {i}')
```
