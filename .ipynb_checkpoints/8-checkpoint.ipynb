```python
class Node(object):
    def __init__(self, stringList, parentNode=None):
        self.header1 = int(stringList[0])
        self.header2 = int(stringList[1])
        self.toProc = stringList[2:]
        self.parentNode = parentNode
        self.childMetaSum = 0
        self.childNode = []
        
        self.proc()
        
    
    def proc(self):
        if self.header1 == 0:
            if len(self.toProc) == self.header2:
                self.metaData = self.toProc
                self.sumMetaMethod()
                if self.parentNode:
                    self.parentNode.gettingChildMetaSum(self.sumMeta, True)
                else:
                    self.wrappingUp()
            else:
                self.metaData = self.toProc[:self.header2]
                self.toProcSib = self.toProc[self.header2:]
                self.sumMetaMethod()
                if self.parentNode:
                    self.parentNode.gettingChildMetaSum(self.sumMeta, done=False)
                    self.parentNode.createChild(self.toProcSib)
                
        else:
            self.createChild(self.toProc)
    
    def createChild(self, childToProc, lastToBeCreated=False):
        if len(self.childNode) < self.header1:
            if lastToBeCreated == True:
                self.metaData = childToProc[-self.header2:]
                self.createChild(childToProc[:-self.header2])
            if self.header1 - len(self.childNode) == 1:
                self.createChild(childToProc, True)
            self.childNode.append(Node(self.toProc, self))
        else:
            if len(childToProc) == self.header2:
                self.metaData = childToProc
                self.sumMetaMethod()
                print(self.sumMeta+self.childMetaSum)
                
            else:
                self.metaData = childToProc[:self.header2]
                self.sumMetaMethod()
                self.parentNode.createChild(childToProc[self.header2:])
                
            
        
    def sumMetaMethod(self):
        self.sumMeta = 0
        for meta in self.metaData:
            self.sumMeta += int(meta)
            
    def getWrapUp(self, lastChildTotal):
        self.childMetaSum += lastChildTotal
        self.wrappingUp()
            
            
    def wrappingUp(self):
        self.sumMetaMethod()
        self.total = self.sumMeta + self.childMetaSum
        if self.parentNode:
            self.parentNode.getWrapUp(self.total)
        else:print(self.total)

    
    def gettingChildMetaSum(self, childMetaSum:int, done:bool):
        self.childMetaSum += childMetaSum
        if done:
            self.wrappingUp()
```
