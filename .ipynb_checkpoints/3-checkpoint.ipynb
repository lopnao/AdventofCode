## Methods

```python
def creatingArray(x:int, y:int):
    return np.zeros((x,y), dtype=int)
```

```python
def imprintSquare(baseArray, claimList, claimTemp, leftSide:int, topSide:int, width:int, height:int, claimID:int):
    for x in range(width):
        for y in range(height):
            if baseArray[leftSide+x, topSide+y] == 0:
                baseArray[leftSide+x, topSide+y] = claimID
                if claimID not in claimList and claimID not in claimTemp:
                    claimList.add(claimID)
            else:
                if int(baseArray[leftSide+x, topSide+y]) in claimList:
                    claimList.remove(int(baseArray[leftSide+x, topSide+y]))
                    claimTemp.add(int(baseArray[leftSide+x, topSide+y]))
                if claimID in claimList:
                    claimList.remove(claimID)
                    claimTemp.add(claimID)
                baseArray[leftSide+x, topSide+y] = 999999
```

```python
def procSquareInput(string: str, baseArray):
    res = re.search('#(\d+) @ (\d+),(\d+): (\d+)x(\d+)', string)
    claimID = int(res.group(1))
    leftSide = int(res.group(2))
    topSide = int(res.group(3))
    width = int(res.group(4))
    height = int(res.group(5))
    imprintSquare(baseArray=baseArray,claimList=claimList, claimTemp=claimTemp, leftSide=leftSide, topSide=topSide, 
                  width=width, height=height, claimID=claimID)
```

```python
def procData(data:str, baseArray):
    datas = data.splitlines()
    for data in datas:
        procSquareInput(string=data, baseArray=baseArray)
```
