```python
import getres
import re
import matplotlib.pyplot as plt 
import collections
```

```python
class Star(object):
    def __init__(self, posX, posY, velX, velY):
        self.posX = self.posXO = posX
        self.posY = self.posYO = posY
        self.velX = velX
        self.velY = velY

        
    def timeup(self, t):
        self.posX = self.posXO
        self.posY = self.posYO
        if self.velX >= 0:
            self.posX += (self.velX * t)
        else:
            self.posX -= (abs(self.velX) * t)
        if self.velY >= 0:
            self.posY += (self.velY * t)
        else:
            self.posY -= (abs(self.velY) * t)
        
```

```python
def getPos(d):
    mo = re.search('position=< ?(-?\d+), +?(-?\d+)> velocity=< ?(-?\d+), +?(-?\d+)>', d)
    posX, posY, velX, velY = int(mo.group(1)), int(mo.group(2)), int(mo.group(3)), int(mo.group(4))
    return posX, posY, velX, velY
```

```python
print(getPos('position=< 7,  0> velocity=<-1,  0>'))
```

```python
def getStars(data):
    data = data.splitlines()
    return [Star(posX=getPos(d)[0], posY=getPos(d)[1], velX=getPos(d)[2], velY=getPos(d)[3]) for d in data]
```

```python
def plotStars(starList, t=0, plot=False):
    contacts = 0
    grid = collections.deque()
    for s in starList:
        s.timeup(t)
        grid.append((s.posX, s.posY))
        if plot:
            plt.plot(s.posX, s.posY, 'ro')
    for s in starList:
        if (s.posX, s.posY+1) in grid and (s.posX, s.posY-1) in grid:
            contacts += 1
    if plot:
        plt.gca().invert_yaxis()
        plt.show()
    return contacts
```

```python
data = getres.getInput(10)
```

```python
starList = getStars(data)
```

```python
for t  in range(1,20000,1):
    if plotStars(starList, t):
        print('found with t =', t)
```

```python
for t  in range(10140,10149):
    plotStars(starList, t, plotStars)
```
